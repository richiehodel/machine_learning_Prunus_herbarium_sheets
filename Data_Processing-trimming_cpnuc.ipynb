# This loop will take in a directory of images, and trim the top and bottom of the images

```python
cd /content/drive/MyDrive/DL_images/CP_NUC/model_dev/solitary_corymbose/
```

```python
sc_list = os.listdir()
```

```python
len(sc_list)
```

```python
for y in range(0,len(sc_list)):
  img_hold = '/content/drive/MyDrive/DL_images/CP_NUC/model_dev/solitary_corymbose/'+sc_list[y]
  my_img = cv2.imread(img_hold)
  ax0=np.shape(my_img)[0]
  ax1=np.shape(my_img)[1]
  ax2=np.shape(my_img)[2]
  ax_mod_bottom=int(0.8*ax0)
  ax_mod_top=int(0.2*ax0)
  keeper = my_img[ax_mod_top:ax_mod_bottom,0:ax1,:]
  outfile = '/content/drive/MyDrive/DL_images/CP_NUC/model_dev/SC_trim1/'+sc_list[y]
  cv2.imwrite(outfile,keeper)
```
